```python
import keras
import nltk
import re
import numpy as np
import pandas as pd

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
from keras.preprocessing.text import one_hot
from keras.utils import to_categorical
from keras.preprocessing.sequence import pad_sequences
from sklearn.model_selection import train_test_split

from nltk.stem import WordNetLemmatizer
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import Flatten
from keras.layers import Dropout
from keras.layers.embeddings import Embedding
from keras.layers.convolutional import Conv1D
from keras.layers.convolutional import MaxPooling1D
from sklearn.feature_extraction.text import TfidfVectorizer

nltk.download('stopwords')
stopwords = set(stopwords.words('portuguese'))

import warnings
warnings.filterwarnings("ignore")
```

```python
gustavo_lima = [letra1, letra2, letra3, letra4, letra5]
michel_telo = [letra6, letra7, letra8, letra9, letra10]
```

```python
lemmatizer = WordNetLemmatizer() 

def text_basics(text):
    text = text.lower()
    text = re.sub(r'[\W\s]', ' ', text)
    text = " ".join([word for word in word_tokenize(text) if word not in stopwords])
    text = " ".join([lemmatizer.lemmatize(word) for word in word_tokenize(text)])
    return text
```

```python
def split_sentences(text):
    sentences = text.split("\n")
    sentences = [text_basics(text) for text in sentences if len(text) > 3]
    return sentences
```

```python
data = []
for letra in gustavo_lima:
    for frase in split_sentences(letra):
        data.append((frase, "Gustavo Lima"))
for letra in michel_telo:
    for frase in split_sentences(letra):
        data.append((frase, "Michel Teló"))
```

```python
df = pd.DataFrame(data, columns=["texto", "cantor"])
cantores_dummies = pd.get_dummies(df["cantor"])
cantores = np.unique(df["cantor"])
df = df.drop(["cantor"], axis=1)
df = pd.concat([df, cantores_dummies], axis=1)
df
```

```python
y = df[["Gustavo Lima", "Michel Teló"]].values
textos = df["texto"].values
```

```python
# Cria os vetores one_hot
vocab_size = 50
encoded_docs = [one_hot(d, vocab_size) for d in textos]
print(encoded_docs[:10])
```

```python
# Define os documentos com o máximo de palavras existente nos textos
max_length = max([len(d) for d in encoded_docs])
X = pad_sequences(encoded_docs, maxlen=max_length, padding='post')
```

```python
# Divide o dataset em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, shuffle=True)
```

```python
# Define os parâmetros da rede
epochs = 50
batch_size = 1
num_class = y.shape[1]
filtros = 12 # Numero de kernels da CNN
tamanho_kernel = 3 # tamanho do kernel
linhas = X_train.shape[0]  # Número de linhas do dataset de treino
n_features = X_train.shape[1] # Número máximo de palavras no dataset geral
```

```python
#Monta o modelo
model = Sequential()
# Cria uma camada de 50 neurônios de entrada, com 8 de saída
# e para cada entrada deve ter como dimensões o número de dimensões do vetor de embeddings
model.add(Embedding(input_dim=vocab_size, output_dim=8, input_length=max_length))
# Cria uma camada convolucional de 1 dimensão, tendo 12 kernels de tamanho 3
# utilizando relu como ativação 
model.add(Conv1D(filters=filtros, kernel_size=tamanho_kernel, activation='relu'))
# Descarta 20% dos resultados para evitar overfit
model.add(Dropout(0.3))
# Faz o pooling de 5 em 5
model.add(MaxPooling1D(pool_size=5))
# Achata a matriz
model.add(Flatten())
# Cria uma camada densa, com 4 neurônios de entrada
model.add(Dense(4, activation='relu'))
# Cria uma camada de saída, com 2 neurônios e ativação softmax para retornar o percentual
model.add(Dense(num_class, activation='softmax'))
model.compile(loss='mean_squared_error', optimizer='adam', metrics=['accuracy'])
```

```python
# Faz o fit
model.fit(X_train, y_train, epochs=epochs, batch_size=batch_size, verbose=1)# Avalia o modelo
loss, accuracy = model.evaluate(X_test, y_test, batch_size=batch_size, verbose=0)
print(model.metrics_names)
print("Accuracy: ", accuracy)
print("Loss: ", loss)
```

```python
def text_predict(texto):
    texto = text_basics(texto)
    print(texto)
    values = one_hot(texto, vocab_size)
    X_input = pad_sequences([values], maxlen=max_length, padding='post')
    print("X values:", X_input[0])
    y_predict = list(model.predict(X_input)[0])
    print("predict values:", y_predict)
    max_value = max(y_predict)
    return cantores[y_predict.index(max_value)]
```
